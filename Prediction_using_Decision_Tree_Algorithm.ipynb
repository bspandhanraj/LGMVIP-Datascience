```python
from sklearn.datasets import load_iris 

import pandas as pd 

from sklearn.model_selection import train_test_split 

from sklearn.tree import DecisionTreeClassifier 

import matplotlib.pyplot as plt 

from sklearn.metrics import accuracy_score 

from sklearn.tree import plot_tree 

from sklearn import tree 
```

```python
data = load_iris() 

data 
```

```python
df = pd.DataFrame(data.data, columns = data.feature_names) 

df['target'] = data.target 
```

```python
name = [x for x in df.columns if x!="target"] 

x= df[name] 

y= df["target"] 
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y) 
```

```python
DTC = DecisionTreeClassifier()  

DTC.fit(X_train,Y_train) 
```

```python
predictions = DTC.predict(X_test) 
```

```python
simplified = pd.DataFrame({"Actual Values":Y_test,"Predicted Values":predictions}) 

simplified.head() 
```

```python
accuracy_score(Y_test,predictions) 
```

```python
Flower = ["Sepal length", "Sepal width", "Petal length", "Petal width"] 

names = ["Setosa", "Versicolor", "Virginica"] 

plot_tree(DTC,feature_names = Flower, class_names = names, filled = True) 
```

```python
DTC2= DecisionTreeClassifier().fit(x,y) 
```

```python
plt.figure(figsize = (20,15)) 

tree = tree.plot_tree(DTC2, feature_names = Flower, class_names = names, filled = True) 
```
